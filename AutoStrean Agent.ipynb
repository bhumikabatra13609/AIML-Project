```python
import json

# Load knowledge base
with open("knowledge_base.json", "r") as file:
    kb = json.load(file)

# Lead capture function
def mock_lead_capture(name, email, platform):
    print("\n✅ Lead captured successfully:")
    print(name, email, platform)

# Memory
user_data = {"name": None, "email": None, "platform": None}

# Intent detection
def detect_intent(text):
    text = text.lower()

    if "hi" in text or "hello" in text:
        return "greeting"
    elif "price" in text or "plan" in text:
        return "pricing"
    elif "buy" in text or "try" in text or "want" in text:
        return "high_intent"
    else:
        return "general"

# Chat loop
while True:
    user = input("\nYou: ")
    intent = detect_intent(user)

    if intent == "greeting":
        print("Bot: Hello! Welcome to AutoStream 😊")

    elif intent == "pricing":
        print("Bot: We have two plans:")
        print("Basic - $29/month (10 videos, 720p)")
        print("Pro - $79/month (Unlimited videos, 4K, AI captions)")

    elif intent == "high_intent":
        print("Bot: Great! Let's get you started.")

        user_data["name"] = input("Enter Name: ")
        user_data["email"] = input("Enter Email: ")
        user_data["platform"] = input("Enter Platform (YouTube/Instagram): ")

        mock_lead_capture(
            user_data["name"],
            user_data["email"],
            user_data["platform"]
        )
        break

    else:
        print("Bot: Please ask about pricing or plans.")
```
